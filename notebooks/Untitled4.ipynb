```python
import os
import mlflow.pyfunc
import pandas as pd

mlflow.set_tracking_uri("http://mlflow:5000")

# Tạo thư mục tmp nếu chưa có
os.makedirs("/tmp/mlruns", exist_ok=True)

# Load model vào dst_path
model = mlflow.pyfunc.load_model(
    "models:/Customer_Spending_Prediction_XGB/Production",
    dst_path="/tmp/mlruns"
)

# Tạo dữ liệu test giả lập
data = pd.DataFrame({
    "total_spend": [500, 1000],
    "years_active": [2, 5],
    "n_years_with_orders": [1, 3],
    "avg_order_value": [200, 350],
    "std_order_value": [50, 80],
    "avg_items_per_order": [3, 5],
    "total_orders": [10, 15],
    "age_group": ["25-34", "35-44"],
    "gender": ["M", "F"],
    "education": ["Bachelor", "Master"],
    "income": ["Medium", "High"],
    "race": ["White", "Asian"],
    "state": ["CA", "NY"],
    "region": ["West", "East"],
    "most_freq_category": ["Electronics", "Clothing"]
})

# Dự đoán
predictions = model.predict(data)
print(predictions)
```
