```python
from minio import Minio
import os

# Kết nối tới MinIO
client = Minio(
    "minio:9000",              # hoặc 'minio:9000' nếu cùng network Docker
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Tên bucket
bucket_name = "bronze"

# Tạo bucket nếu chưa có
if not client.bucket_exists(bucket_name):
    client.make_bucket(bucket_name)
    print(f"Bucket '{bucket_name}' created.")
else:
    print(f"Bucket '{bucket_name}' already exists.")

# Thư mục chứa file CSV
folder_path = "/home/jovyan/notebooks/dataverse_files"


# Upload từng file CSV
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        client.fput_object(bucket_name, f"ecommerse/{filename}", file_path)
        print(f"Uploaded {filename} to bucket '{bucket_name}'")

print("All files uploaded successfully!")
```
